```python
base_src = "./drive/MyDrive/Python/machine_learning"
friend_src = base_src + "/friend.csv"
df = pd.read_csv(friend_src)
df.head()
```

```python
# index 2번에 해당 row 가져오기
df.iloc[2]
```

```python
# column job 에 해당되는 데이터 가져오기
df['job']
```

```python
# column job에 해당되는 데이터 가저오기(2)
df.loc[:,'job']
```

```python
# 슬라이싱 기능을 통해 여러 행 가져오기
df.iloc[[2,3]] # 슬라이싱 개념을 쓰지 않았다.
```

```python
# 슬라이싱 개념 사용
df.iloc[:6] # 0~5
```

```python
# "많이 하는 질문!"
'''iloc VS loc 뭐가 다른거죠???
- iloc는 인덱스와 컬럼을 리스트 배열로 선택하는 것!
- loc는 인덱스와 컬럼을 문자로 선택하는 것!
- 따라서 상황에 맞게 선택하는 것이 중요!'''
```

```python
# 조건 필터링 가져오기
# 30대 이상만 가져오기
df[df['age']<30]
```

```python
## job이 intern인 사람 가져오기
df[df['job'] == 'intern']
```

```python
## 조건 여러개
## 30대 이상 40대 이하 (and function)
df[(df['age']>=30) & (df['age']<=40) & (df['age']>20)]
# 조건 여러개 공식 : df를 쓰고 -> 데이터프레임 []쓰고 -> 데이터프레임 안에 ()를 사용
# 내가 조건을 넣고 싶은 것을 또 ()씌우고 & and function을 사용한다.
```

```python
## 조건 여러개 30대 미만 혹은 40대 초과 (or function)
df[(df['age']<30) | (df['age']>40)]
```

```python
# in을 통한 포함 조건 걸기
df[(df['job']=='student') | (df['job']=='manager')]
# df[df['job'] in ['student','manager']]
```

```python
# DataFrame 행, 열 삭제
import numpy as np
df = pd.DataFrame(np.arange(12).reshape(3, 4),
                  columns=['A', 'B', 'C', 'D'])
df
```

```python
# B , C 컬럼을 삭제하고 싶습니다.
df.drop(["B","C"],axis=1,inplace=True)
```

```python
df
```

```python
# index 0과 1을 삭제하겠습니다.
# axis는 왜 안넣나요? axis는 Default 값이 0이다.
# axis=0을 넣어도 상관은 없다.
# axis=1을 넣게 되면 Error가 나온다. 왜? column에는 index 0과1이 없기 때문이다.
df.drop([0,1],axis=0)
```
