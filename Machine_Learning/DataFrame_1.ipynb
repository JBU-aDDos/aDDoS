```python
# iloc 처럼 결을 맞춰 정형화되게 가져오기 => loc는 문자로 검색
df.loc[:,'job'] # 왼쪽은 row, 오른쪽은 colum
```

```python
df.iloc[:,:2]
# row만 하고싶을때는 오른쪽에 :를 빼줘야한다.
# colum만 하고싶을때는 왼쪽에 :을 무조건 넣어줘야한다.
```

```python
# 문자 => 실수 => 정수
# "많이 하는 질문!"
'''iloc VS loc 뭐가 다른거죠???
- iloc는 인덱스와 컬럼을 리스트 배열로 선택하는 것!
- loc는 인덱스와 컬럼을 문자로 선택하는 것!
- 따라서 상황에 맞게 선택하는 것이 중요!'''
```

```python
df = pd.DataFrame(np.arange(12).reshape(3,4),
                  columns=['A','B','C','D'])
```

```python
df.drop(['B','C'],axis=1,inplace=True)
```

```python
df.drop([0,1],axis=0)
```

```python
# 데이터 수정
df.loc[1,'C'] = '육'
```

```python
df.iloc[1,2] = '6'
```

```python
df
```

```python
# 슬라이싱의 기능을 활용하면 내가 원하는 범위까지 잘 바꿀 수 있다.
df.iloc[:,1] = "끝~!"
```

```python
df.iloc[:,:] = "오늘 수업 시작!"
```
